# Question 1
### users data

```python
# step 1:

import numpy as np
import pandas as pd
```

```python
# step 2 and step 3:

users = pd.read_csv(r"https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user",sep="|")
users.head()
```

```python
# step 4:

users.columns
```

```python
mean_age_per_occupation = users.groupby('occupation').age.mean()

mean_age_per_occupation
```

```python
# step 5:

total_count = users.groupby("occupation").gender.count()
male_count = users.where(users.gender=="M").groupby(["occupation","gender"]).gender.count()
ratio = (male_count / total_count).sort_values(ascending = False)

ratio
```

```python
# step 6:

min_max_per_occupation = users.groupby(["occupation"]).age.agg(["min","max"])

min_max_per_occupation
```

```python
# step 7:

mean_age_genderwise_for_each_occupation = users.groupby(["occupation","gender"]).age.mean()

mean_age_genderwise_for_each_occupation
```

```python
# step 8:

total_count = users.groupby('occupation').gender.agg(['count'])
male_count = users.where(users.gender=='M').groupby(['occupation','gender']).gender.agg(['count'])
female_count = users.where(users.gender=='F').groupby(['occupation','gender']).gender.agg(['count'])

male_percent = (male_count / total_count) * 100
female_percent = (female_count / total_count) * 100

male_and_female_percent_each_occupation = pd.merge(male_percent, female_percent, on='occupation')

male_and_female_percent_each_occupation
```

# Question 6

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
marriage_data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
marriage_data.head()
```

```python
year_of_marriage = marriage_data.Year.values

marriages_per_1000 = marriage_data.Marriages_per_1000.values

divorces_per_1000 = marriage_data.Divorces_per_1000.values

plt.plot(year_of_marriage, marriages_per_1000, color="green")
plt.plot(year_of_marriage, divorces_per_1000, color="red")
plt.xlabel("Number of Years")
plt.ylabel("Mariages and divoces per 1000")
plt.title("Pictorial representation of number of marriages and divorces in the U.S.A. between 1867 and 2014.")
plt.show()
```

# Question 7

```python
years_1900 = marriage_data.Year >= 1900
years_2000 = marriage_data.Year <= 2000
data_1900_2000 = marriage_data[years_1900 & years_2000]

plt.bar(data_1900_2000['Year'], data_1900_2000['Marriages_per_1000'],color="grey")
plt.bar(data_1900_2000['Year'], data_1900_2000['Divorces_per_1000'],color="black")

plt.title("Total number of Marraiges and Divorces per capita in USA during 20th centuary.")
plt.xlabel("Between 1900 and 2000")
plt.ylabel("Marriages & Divorces per capita in US")
plt.show()
```

# Question 9

```python
roman_emperor_data = pd.read_csv('roman-emperor-reigns.csv')
roman_emperor_data.head()

lower_data = roman_emperor_data.where(roman_emperor_data.Cause_of_Death=="Assassinated").Cause_of_Death.count()
remaining = roman_emperor_data.Cause_of_Death.count() - lower_data

labels = ["Other Causes of Deaths","Assassinated"]
plt.pie([remaining, lower_data],labels=labels,autopct='%.2f%%')
plt.title("Deaths in Roman Emperor")

plt.show()
```

```python
labels = list(roman_emperor_data.groupby("Cause_of_Death").groups)
data = list(roman_emperor_data.groupby("Cause_of_Death").count()['Emperor'])

plt.pie(data , labels=labels,autopct='%.2f%%')
plt.title("Deaths in Roman Emperor")

plt.show()
```
