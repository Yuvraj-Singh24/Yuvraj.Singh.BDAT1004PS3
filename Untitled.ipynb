# Question 1
### users data

```python
# step 1:

import numpy as np
import pandas as pd
```

```python
# step 2 and step 3:

users = pd.read_csv(r"https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user",sep="|")
users.head()
```

```python
# step 4:

users.columns
```

```python
mean_age_per_occupation = users.groupby('occupation').age.mean()

mean_age_per_occupation
```

```python
# step 5:

total_count = users.groupby("occupation").gender.count()
male_count = users.where(users.gender=="M").groupby(["occupation","gender"]).gender.count()
ratio = (male_count / total_count).sort_values(ascending = False)

ratio
```

```python
# step 6:

min_max_per_occupation = users.groupby(["occupation"]).age.agg(["min","max"])

min_max_per_occupation
```

```python
# step 7:

mean_age_genderwise_for_each_occupation = users.groupby(["occupation","gender"]).age.mean()

mean_age_genderwise_for_each_occupation
```

```python
# step 8:

total_count = users.groupby('occupation').gender.agg(['count'])
male_count = users.where(users.gender=='M').groupby(['occupation','gender']).gender.agg(['count'])
female_count = users.where(users.gender=='F').groupby(['occupation','gender']).gender.agg(['count'])

male_percent = (male_count / total_count) * 100
female_percent = (female_count / total_count) * 100

male_and_female_percent_each_occupation = pd.merge(male_percent, female_percent, on='occupation')

male_and_female_percent_each_occupation
```

### Wind Statistics

```python
# step 1:

import numpy as np
import pandas as pd
```

```python
# step 2:

data = pd.read_csv("wind.txt", sep='\s+')
data.head()
```

```python
# step 3:

data.columns
```

```python
data["Date"] = pd.to_datetime(data[["Yr","Mo","Dy"]].astype(str).agg('-'.join, axis = 1))
data = data.drop(columns=["Yr","Mo","Dy"])
data.head()
```

```python
# step 4:

data["Date"] = np.where(pd.DatetimeIndex(data["Date"]).year < 2021, data.Date,data.Date - pd.offsets.DateOffset(years=100))

data
```

```python
# step 5:

correct_data = data.set_index("Date")
correct_data.index.astype("datetime64[ns]")
```

```python
# step 6:

data.isnull().values.sum()
```

```python
# step 7:

correct_data.count().sum()
```

```python
# step 8:

correct_data.mean().mean()
```

```python
# step 9 and step 10:

max_data = correct_data.max()
min_data = correct_data.min()

mean_data = correct_data.mean()
std_data = correct_data.std()

stats = [min_data, max_data, mean_data, std_data]

index = ["Min","Max","Mean","Std"]
loc_stats = pd.DataFrame(stats, index)

loc_stats
```

```python
# step 11:

january_data = correct_data[correct_data.index.month == 1]
january_data.mean()
```
